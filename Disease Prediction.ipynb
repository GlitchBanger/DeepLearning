# Model to predict diseases based on symptoms.



```python
import pandas as pd

dataset = pd.read_csv('Training.csv')
dataset
```

```python
dataset.prognosis.unique()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

le.fit(dataset['prognosis'])
```

```python
dataset['label'] = le.transform(dataset.prognosis)
dataset['label']
```

```python
le.get_params()
```

```python
label_dict = {}
for i in dataset.prognosis.unique():
    label_dict[i] = le.transform([i])[0]
label_dict
```

```python
dataset
```

```python
data_cols = dataset.columns
data_cols = data_cols.tolist()
data_cols.remove('Unnamed: 133')
data_cols.remove('prognosis')
data_cols.remove('label')

label = 'label'
data_cols
```

```python
import numpy as np

data = dataset[data_cols].to_numpy().reshape((-1,1,132))

from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()

classes = dataset['prognosis']

classes = classes.to_numpy()

classes = classes.reshape(-1,1)
label = encoder.fit_transform(classes).toarray().reshape((-1,1,41))

print(label.shape)
data.shape
```

```python
import tensorflow as tf

from tensorflow.keras import layers, Input, Sequential
```

```python
all_ds = tf.data.Dataset.from_tensor_slices((data, label))
print(all_ds)
```

```python
dataset_size = len(all_ds)
train_size = int(0.8 * dataset_size)
val_size = dataset_size - train_size

train_size, val_size
```

```python
all_ds = all_ds.shuffle(10000)

train_ds = all_ds.take(train_size)
val_ds = all_ds.skip(train_size)

len(train_ds), len(val_ds)
```

```python
model = Sequential([
    Input(132),
    layers.Dense(200, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(150, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(100, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(75, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(50, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(41, activation='sigmoid')
])

model.summary()
```

```python
model.compile(optimizer='adam', loss='CategoricalCrossentropy', metrics=['accuracy'])
```

```python
model.fit(train_ds, validation_data=val_ds, epochs=10, batch_size=32)
```

```python
for d, l in all_ds.take(10):
    print('Prediction: ', encoder.inverse_transform(model.predict(d).round())[0][0])
    print('Actual: ', encoder.inverse_transform(l)[0][0])
```

```python
test_dataset = pd.read_csv('Testing.csv')
test_dataset
```

```python
test_data = test_dataset[data_cols].to_numpy().reshape((-1,1,132))

test_data.shape
```

```python
test_label = encoder.transform(test_dataset['prognosis'].to_numpy().reshape((-1,1))).toarray().reshape((-1,1,41))
test_label.shape
```

```python
test_ds = tf.data.Dataset.from_tensor_slices((test_data, test_label))
test_ds
```

```python
model.evaluate(test_ds)
```
